# Feature Engineering


```python
# Creating new columns
dfcpy['installs_per_size'] = dfcpy['Installs'] / dfcpy['Size']
```

```python

# Analyzing the Last update
# Recency of update
# In this we'll be extracting some features from the original column df['Last_Updated']

lasUpdated = df['Last Updated']

lasUpdated = pd.to_datetime(lasUpdated, errors='coerce')

reference_date = lasUpdated.max()

# days since last update
dfcpy['Recency_of_Update'] = (reference_date - lasUpdated).dt.days
```

```python
#Group Category and Genres
dfcpy.info()
dfcpy['Cat_Gen'] = dfcpy.groupby('Category')['Genres'].count()
dfcpy['Cat_Gen'].fillna(0, inplace=True)
dfcpy
```

```python
#Interaction Features
dfcpy['ReviewsPerInstall'] = dfcpy['Reviews'] / dfcpy['Installs']
dfcpy
```

```python
y = dfcpy['Reviews']
dfcpy.drop(columns=['Reviews'], inplace=True)
```

```python
from sklearn.model_selection import train_test_split
X_train, x_test, Y_train, y_test = train_test_split(dfcpy, y, test_size=0.2, random_state=100)
```

# Hyper Parameter Tuning

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeRegressor # Keep DecisionTreeRegressor

# Define the parameter grid for GridSearchCV
param_grid = {
    'max_depth': [3, 5, 7, 10, 15], # Added more depth options
    'min_samples_split': [2, 5, 10], # Minimum number of samples required to split an internal node
    'min_samples_leaf': [1, 2, 4], # Minimum number of samples required to be at a leaf node
    'max_features': [None, 'sqrt', 'log2'] # Number of features to consider when looking for the best split
}

# Initialize GridSearchCV with DecisionTreeRegressor
gscv = GridSearchCV(estimator=DecisionTreeRegressor(random_state=100), # Added random_state for reproducibility
                    param_grid=param_grid,
                    scoring='neg_mean_squared_error', # Using negative MSE for optimization
                    cv=3, # Using 3-fold cross-validation
                    n_jobs=-1) # Use all available cores
```

```python
bestModel = gscv.fit(X_train, Y_train)
```

```python
bestModel.best_params_
```

# Machine Learning Technique

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, r2_score, classification_report, confusion_matrix
```

```python
bestParams = bestModel.best_params_
bestParams
```

```python
#Creating a new RandomForestRegressor instance with best hyperParams
finalModel = DecisionTreeRegressor(random_state=100, **bestParams)

#Train the final model
finalModel.fit(X_train, Y_train)
```

```python
yPred = finalModel.predict(x_test)
```

```python
# Evaluation metrics

print("R-squared:", r2_score(y_test, yPred))
```
